# data_2_1.csv, data_2_2.csv, data_2_3.csv  merging and clean the result datafram  

```python
# merging dataset data_2_1 and data_2_2
data_2_1_df = pd.read_csv('data/data_2_1.csv')
data_2_2_df = pd.read_csv('data/data_2_2.csv')

data_3_df = pd.merge(data_2_1_df, data_2_2_df, on='API')
# data_3_df.head()
```

```python
data_2_3_df = pd.read_csv('data/data_2_3.csv')
```

```python
# merge data_3 and data_2_3
welldata02_df = pd.merge(data_3_df, data_2_3_df, on='API')
welldata02_df.head()
```

```python
# get the correct form of APT
welldata02_df['API'] = welldata02_df['API'].str.replace('-','').str.replace('0000','')
welldata02_df['API'] = pd.to_numeric(welldata02_df['API'])

# change HORIZANTAL TO H & VERTICAL TO V IN 'Wellbore Profile - Aliased' column 
welldata02_df['Wellbore Profile - Aliased'] = welldata02_df['Wellbore Profile - Aliased'].str.replace('HORIZONTAL','H').\
str.replace('VERTICAL','V')


```

```python
welldata02_df.head()
```

```python
list(welldata02_df)
```

```python
welldata02_df.rename(columns={             
 
 'Name': 'Lease_Name',
 'Well Type - Aliased': 'Production_type',
 'Status - Aliased': 'Production_Status',
 'Wellbore Profile - Aliased':'Drilling_Type' ,
 'First Production Date':'First_Production_Day',
 'CUM Oil':'Cum_Oil',
 'CUM Gas':'Cum_Gas',
 'CUM Water':'Cum_water',
 'Latitude':'Lat',
 'Longitude':'Long',
 'Spud Date':'Spud_Date',
 'Completion Date':'Completion_Date',
 'Well Number':'Well_Number',
 'Upper Perf':'Upper_Perforation',
 'Lower Perf':'Lower_Perforation',
 'Formation':'Reservoir',
 'operator':'Operator',
 'CUM BOE':'Cum_BOE'}
                     
                     
                     , inplace=True)
```

```python
# welldata02_df
# welldata02_df.dropna(axis = 1)
welldata02_df.head()
```

```python
# droping unecessary rows with weired values !! 
welldata02_df = welldata02_df[welldata02_df.Drilling_Type != 'DIRECTIONAL']
welldata02_df = welldata02_df[welldata02_df.Production_type != 'DIRECTIONAL']
```

```python
welldata02_df.to_csv('data/part2_wellData_Cleaned.csv')
```

```python
# concatenate welldata01 and welldata02 to obtain the finall dataset 

frames = [welldata01_df,welldata02_df]
welldata_df = pd.concat(frames)

```
